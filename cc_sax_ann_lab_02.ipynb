# Neural net ensemble for sax
### Step 2
### Make an ensemble of ANNs and save them to disk

```python
import os
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.metrics import Recall

from _all_saxes import x_train, x_test, y_train, y_test
```

```python
# Make an ensemble of ANNs

VALIDATION_SIZE = 0.2
EPOCHS = 30
BATCH_SIZE = 16
# RECALL_THRESHOLD = 0.85

h1_values = [24, 28, 32]
d1_values = [0.1, 0.15]
h2_values = [8, 12, 16]
d2_values = [0.1]
h3_values = [8, 12, 16]

for h1_value in h1_values:
    for d1_value in d1_values:
        for h2_value in h2_values:
            for d2_value in d2_values:
                for h3_value in h3_values:
                    fn ='model_sax_{}_{}_{}_{}_{}.h5'.format(h1_value,
                                                             d1_value,
                                                             h2_value,
                                                             d2_value,
                                                             h3_value)
                    if os.path.isfile('./keras_models/' + fn):
                        # print(fn + " EXISTS")
                        continue
                    # print("Generating " + fn)       
                    model = Sequential()
                    model.add(Dense(h1_value, activation='relu'))
                    model.add(Dropout(d1_value))
                    model.add(Dense(h2_value, activation='relu'))
                    model.add(Dropout(d2_value))
                    model.add(Dense(h3_value, activation='relu'))
                    model.add(Dense(1, activation='sigmoid'))
                    model.compile(loss='binary_crossentropy', optimizer='adam', 
                                  metrics=[Recall()])
                    model.fit(x_train, y_train, epochs=EPOCHS, batch_size=BATCH_SIZE, 
                              validation_split=VALIDATION_SIZE)
                    model.save('./keras_models/' + fn)
```
